# Tree Ensemble

- Trees are highly sensitive if there is a small change in the data.
- In tree ensemble, we will make a possible tree from the data, and we will choose each tree prediction and vote on who has the most votes.


- For that, we will use sampling with the replacement method.


- In order to perform the tree assembly, we need a technique called sampling with replacement.


- We will make a new training set using the sampling method, choose a random row, and select whether it can be duplicated, and so on.

## Random Forest

In a random forest, we will make a random subset of the dataset, make a decision tree of all the random subsets, predict from them, and then use voting again.

## XGBoost

Here we will sample with replacement to create a training set, but instead of picking from all examples with equal 1/m probability, it will be more likely to pick misclassified examples from previously trained trees. Basically,  first we will create a data set like a random subset, make a decision tree, and predict from them. For the next subset, we will choose the misclassified data from that prediction and make new data. This process continues.
