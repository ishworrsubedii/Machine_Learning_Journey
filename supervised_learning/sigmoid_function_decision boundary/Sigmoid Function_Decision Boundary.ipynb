```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Data
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])
w = 0
b = 0
```

```python
#logistci regression 
def sigmoid(x, w, b):
    m = x.shape[0]  # Get the number of samples in x
    logistic = np.zeros((m,))
    for i in range(m):
        z = w * x[i] + b
        logistic[i] = 1 / (1 + np.exp(-z))
        
    return logistic

logistic_calculation = sigmoid(x_train, w, b)
```

```python
logistic_calculation
```

```python
plt.scatter(x_train,y_train,label='Actual Data',c='r')
plt.plot(x_train,logistic_calculation,label='Model Prediction')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

This is just an example of a logistic or sigmoid function; the model prediction is not applicable to the real world. If you want to predict correctly, you have to find or calculate the values of w and b using gradient descent.
