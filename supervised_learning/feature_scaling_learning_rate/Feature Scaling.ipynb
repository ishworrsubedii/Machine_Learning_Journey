
# Feature Scaling

- Mean normalization: $x_i := \dfrac{x_i - \mu_i}{max - min} $ 
- Z-score normalization-$x^{(i)}_j = \dfrac{x^{(i)}_j - \mu_j}{\sigma_j}$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_data = pd.read_csv('houses.txt')
x_features = house_data[['Size', 'Number_of_Bedrooms', 'Number_of_floors', 'Age_of_Home']]
X_train = x_features.values
```

```python
X_min = np.min(X_train,axis=0)   
X_max = np.max(X_train,axis=0) 
X_range = X_max - X_min
X_mean = np.mean(X_train,axis=0)
mean_normalization = (X_train - X_mean) / X_range

sigma = np.std(X_train, axis=0)#Standard Deviation
z_score_normalization = (X_train - X_mean) / sigma #Z score Normalization

fig, ax = plt.subplots(1, 3, figsize=(12, 3))
ax[0].scatter(X_train[:, 0], X_train[:, 3])
ax[0].set_xlabel('Size')
ax[0].set_ylabel('Age_of_Home')
ax[0].set_title("Unnormalized")
ax[0].axis('equal')

ax[1].scatter(mean_normalization[:, 0], mean_normalization[:, 3])
ax[1].set_xlabel('Size')
ax[1].set_ylabel('Age_of_Home')
ax[1].set_title(r"X - $\mu/mix-min$")
ax[1].axis('equal')

ax[2].scatter(z_score_normalization[:, 0], z_score_normalization[:, 3])
ax[2].set_xlabel('Size')
ax[2].set_ylabel('Age_of_Home')
ax[2].set_title(r"Z-score normalized")
ax[2].axis('equal')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
fig.suptitle("Before And After Normaliztion")
plt.show()
```
