# Classification And Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

## One Variable

```python
x_train=np.array([0,1,2,3,4,5])
y_train=np.array([0,0,0,1,1,1])

```

```python
pos=y_train==1
neg=y_train==0


# fig,ax=plt.subplot(1,2,figsize=(8,3))
plt.scatter(x_train[pos],y_train[pos],c='r',marker='X')
plt.scatter(x_train[neg],y_train[neg],c='b',marker='o')
plt.xlabel("x")
plt.ylabel('Y')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
pos = y_train2 == 1
neg = y_train2 == 0

plt.scatter(x_train2[pos, 0], x_train2[pos, 1], c='r', marker='x', label='Positive Class')
plt.scatter(x_train2[neg, 0], x_train2[neg, 1], c='b', marker='o', label='Negative Class')
plt.xlabel("x1")
plt.ylabel('x0')
plt.title("Using Two Variable")

plt.legend()
plt.show()
```

# Linear regression for classification

```python
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([0,1,2,3,4,5])
y=np.array([0,0,0,1,1,1])
x_train=np.array(x).reshape(-1,1)
y_train=np.array(y).reshape(-1,1)

linear_regression=LinearRegression()
```

```python
linear_regression.fit(x_train,y_train)
model_prediction=linear_regression.predict(x_train)
```

```python
pos=y_train==1
neg=y_train==0


# fig,ax=plt.subplot(1,2,figsize=(8,3))
plt.scatter(x_train[pos],y_train[pos],c='r',marker='X')
plt.scatter(x_train[neg],y_train[neg],c='b',marker='o')
plt.plot(x_train,model_prediction)
plt.xlabel("x")
plt.ylabel('Y')
plt.title('Classification Using Linear regr')
plt.show()
```

Linear regression is not applicable for all types of data, so we will have to use logistic regression.
