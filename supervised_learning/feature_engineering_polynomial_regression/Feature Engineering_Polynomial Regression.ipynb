## Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.arange(0, 20, 1) #Data
y_train = 1 + X**2
x_train = X.reshape(-1, 1)
```

```python
#To find the value of b and w we need to calculate gradient descent.

def gradient_descent(x, y, w, b, alpha, num_iters):
    m, n = x.shape
    dj_dw = np.zeros((n,))
    dj_db = 0.
    j_history = []

    for i in range(num_iters):
        for j in range(m):
            xi = x[j,:]
            err = np.dot(xi, w) + b - y[j]
            dj_dw += err * xi
            dj_db += err

        dj_dw /= m
        dj_db /= m
        
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
    return  w, b

w_init = np.zeros((1,))
b_init = 0.
iterations = 1000
alpha = 1e-2

w_final, b_final = gradient_descent(x_train, y_train, w_init, b_init, alpha,iterations)
print(f"Final Value of w : {w_final}, Final Value of b: {b_final:0.2f}")
```

```python
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.title("Polynomial Regression")
plt.plot(x_train,x_train@w_final + b_final, label="Predicted Value")

plt.legend()
plt.show()
```
