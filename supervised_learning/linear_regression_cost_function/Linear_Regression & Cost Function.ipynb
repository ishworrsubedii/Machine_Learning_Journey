__Cost Function__

- Cost function measure different between model predection and the actual values for y.



The equation for cost with one variable is:
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 $$ 
 
where 
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b $$
  
- $f_{w,b}(x^{(i)})$ is our prediction for example $i$ using parameters $w,b$.  
- $(f_{w,b}(x^{(i)}) -y^{(i)})^2$ is the squared difference between the target value and the prediction.   
- These differences are summed over all the $m$ examples and divided by `2m` to produce the cost, $J(w,b)$.  


#### Implementation of linear regression



```python
import numpy as np
import matplotlib.pyplot as plt

w=100
b=100
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
m=len(x_train)


def calculation(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
        
    return f_wb

final_calculation=calculation(x_train,w,b)
plt.title("House Predection")
plt.ylabel('Price')
plt.xlabel('Size')

plt.plot(x_train,final_calculation,label="Model predection",c='b')
#Actual Data
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.legend()
```

## Another Example of different data
