# We can use different activation funcation such as:

## 1) Linear Activation Function
  

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score# Accuracy check for sigmoid function
```

```python
dataset=pd.read_csv('dataset.csv               ')
dataset
```

```python
x_features=['Price','shipping','Marketing','Material']
x=dataset[x_features]
y=dataset[['sold']]
x_train=np.array(x)
y_train=np.array(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_train,y_train,test_size=0.3,random_state=0)
```

```python
model=Sequential([
    Dense(units=3,activation='linear'),#This is the hidden layer of affordance awareness and quality
    Dense(units=1,activation='sigmoid')# This is for output
])
model.compile(loss=MeanSquaredError())# this is to calculate the total of 
model.fit(x_train,y_train,epochs=100)# epochs=Iterations
y_pred_linear=model.predict(x_test)
```

## 2) Sigmoid Activation Function
   - Output(posative/negative)

```python
model=Sequential([
    Dense(units=3,activation='sigmoid'),#This is the hidden layer of affordance awareness and quality
    Dense(units=1,activation='sigmoid')# This is for output
])
model.compile(loss=MeanSquaredError())# this is to calculate the total of 
model.fit(x_train,y_train,epochs=100)# epochs=Iterations
y_pred_sigmoid=model.predict(x_test
                          )
```

accuracy = (number of correct predictions) / (total number of predictions)


```python
accuracy_score(y_test,y_pred_sigmoid.round())
```

## 3) ReLu Activation Function
    (Rectified Linear Unit)
   - Output(0 or posative)
   - f(z) = max(0, z)
   - z=linear function


```python
model=Sequential([
    Dense(units=3,activation='relu'),#This is the hidden layer of affordance awareness and quality
    Dense(units=1,activation='sigmoid')# This is for output
])
model.compile(loss=MeanSquaredError())# this is to calculate the total of 
model.fit(x_train,y_train,epochs=100)# epochs=Iterations
y_pred_relu=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_relu.round())
```

The rectified linear unit (ReLU) function is one of the most commonly used activation functions in neural networks. This is because ReLU is computationally efficient, easy to implement, and has been shown to improve the performance of deep neural networks.
