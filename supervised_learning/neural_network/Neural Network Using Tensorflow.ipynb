# Neural Network Using Tensorflow

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input

from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError


```

```python
dataset=pd.read_csv('dataset.csv')
```

```python
x_features=['Price','shipping','Marketing','Material']
x=dataset[x_features]
y=dataset[['sold']].values
x_train=np.array(x)
y_train=np.array(y)
```

```python
model=Sequential([
    Dense(units=3,activation='sigmoid'),
    Dense(units=1,activation='sigmoid')
]
)

model.compile(loss=MeanSquaredError())
model.fit(x_train,y_train,epochs=100)   #Epochs=Iterations
y_pred = model.predict(x_train)
```

```python

fig,ax=plt.subplots(1,4,figsize=(12,6),sharey=True)
for i in range(len(ax)):
    
    ax[i].scatter(x_train[:,i],y_train,label='Actual data')
    ax[i].set_xlabel(x_features[i])
    ax[i].scatter(x_train[:,i],y_pred,label='Model Prediction',c='r')
ax[0].set_ylabel('Price')
plt.show()
```

```python
w1_1, b1_1 = model.layers[0].get_weights()
print(w1_1,b1_1)
```
